```python
# Objective:
# Analyze how trader performance and risk behavior differ under
# Fear vs Greed market sentiment.
import pandas as pd
import matplotlib.pyplot as plt

master_df = pd.read_csv("../outputs/master_trader_sentiment.csv")

# Keep only rows with valid sentiment
analysis_df = master_df.dropna(subset=["classification"]).copy()

analysis_df.shape
```

```python
# Average PnL by sentiment
analysis_df.groupby("classification")["Closed PnL"].mean()
```

```python
# Win rate by sentiment
analysis_df["is_win"] = analysis_df["Closed PnL"] > 0
analysis_df.groupby("classification")["is_win"].mean()
```

```python
#Risk behavior (position size)
analysis_df.groupby("classification")["Size USD"].mean()
```
